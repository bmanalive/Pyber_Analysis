# Pyber Challenge

### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
rideCount_byCityType = pyber_data_df.groupby(['type']).count()['ride_id']
rideCount_byCityType
```

```python
# 2. Get the total drivers for each city type
sum_driver_count_byCityType = city_data_df.groupby(['type']).sum()['driver_count']
sum_driver_count_byCityType
```

```python
#  3. Get the total amount of fares for each city type
fareSum_byCityType = pyber_data_df.groupby(['type']).sum()['fare']
fareSum_byCityType
```

```python
#  4. Get the average fare per ride for each city type. 
avg_farePerRide = fareSum_byCityType/rideCount_byCityType
avg_farePerRide
```

```python
# 5. Get the average fare per driver for each city type. 
avg_farePerDriver = fareSum_byCityType/sum_driver_count_byCityType
avg_farePerDriver
```

```python
city_type = [pyber_data_df['city'],pyber_data_df['type']]
city_type_df = pd.DataFrame(city_type)
Tcity_type_df = city_type_df.T
type_byCity = Tcity_type_df.groupby(['city']).count()['type']
type_byCity
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_summary_df = pd.DataFrame(
          {"Total Rides": rideCount_byCityType, 
          "Total Drivers": sum_driver_count_byCityType, 
          "Total Fares": fareSum_byCityType,
          "Average Fare per Ride": avg_farePerRide, 
          "Average Fare per Driver": avg_farePerDriver})

# Display the data frame
pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
pyber_summary_df
```

```python
#  8. Format the columns.

# Format the "Total Fares" to have the comma for a thousands separator, a decimal separator and a "$".
pyber_summary_df["Total Fares"] = pyber_summary_df["Total Fares"].map("${:,.2f}".format)

# Format the "Average Fare per Ride" to have a decimal separator and a "$".
pyber_summary_df["Average Fare per Ride"] = pyber_summary_df["Average Fare per Ride"].map("${:.2f}".format)


# Format the "Average Fare per Driver" to have a decimal separator and a "$".
pyber_summary_df["Average Fare per Driver"] = pyber_summary_df["Average Fare per Driver"].map("${:.2f}".format)

pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
sum_fare_byDateAndType = pyber_data_df.groupby(['type','date']).sum()[['fare']]
sum_fare_byDateAndType
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()
sum_fare_byDateAndType = sum_fare_byDateAndType.reset_index()
sum_fare_byDateAndType
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
sum_fare_byDateAndType = sum_fare_byDateAndType.pivot(index='date', columns='type', values ='fare')
sum_fare_byDateAndType
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.

sum_fare_JanThruApril_df = sum_fare_byDateAndType.loc['2019-01-01':'2019-04-29']
sum_fare_JanThruApril_df
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
sum_fare_JanThruApril_df.index = pd.to_datetime(sum_fare_JanThruApril_df.index)
sum_fare_JanThruApril_df.index
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
sum_fare_JanThruApril_df.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
sum_fare_JanThruApril_byWeek_df = sum_fare_JanThruApril_df.resample('W').sum()
sum_fare_JanThruApril_byWeek_df
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 
# Using object-oriented interface method

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

# Create the plot with ax.plt()
fig, ax = plt.subplots(figsize=(16,6))
ax.plot(sum_fare_JanThruApril_byWeek_df)

#Annotate the y-axis label using object-oriented interface method
ax.set_ylabel('Fare ($USD)')

#Annotate the x-axis label using object-oriented interface method
ax.set_xlabel('Date')

#Annotate the Title using object-oriented interface method
ax.set_title("Total Fare by City Type")

#Annotate legend
ax.legend(['Rural','Suburban','Urban'], loc='center')

# Save Figure using object oriented method
plt.savefig("analysis/PyBer_fare_summary_object_oriented.png")

#sum_fare_JanThruApril_byWeek_df.plot()
```
